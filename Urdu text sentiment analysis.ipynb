# removing Noise eg.emoji,url,punctuation,english alphabets , whitespaces etc

```python
import demoji
import re

from urduhack.preprocessing import replace_urls
from urduhack.preprocessing import replace_numbers
from urduhack.preprocessing import remove_punctuation
from urduhack.preprocessing import remove_accents
from urduhack.preprocessing import remove_english_alphabets
from urduhack.preprocessing import normalize_whitespace
from urduhack.normalization import normalize

def clean_urdu_text(urdu_text):

    urdu_text = demoji.replace(urdu_text,"")

    urdu_text = re.sub(r'@[A-Za-z0-9]+', '', urdu_text)

    urdu_text = re.sub(r'#[A-Za-z0-9]*', '', urdu_text)

    urdu_text = replace_urls(urdu_text,replace_with="")

    urdu_text = replace_numbers(urdu_text,replace_with="")

    urdu_text = remove_accents(urdu_text)

    urdu_text = remove_english_alphabets(urdu_text)

    urdu_text = normalize_whitespace(urdu_text)

    urdu_text = remove_punctuation(urdu_text)

    urdu_text = normalize(urdu_text)

    urdu_text = re.sub(r'\+', '', urdu_text)

    urdu_text = re.sub(r'\=', '', urdu_text)

    return urdu_text
```

```python
# Fill missing values with an empty string
train['urdu_text'].fillna('', inplace=True)
valid['urdu_text'].fillna('', inplace=True)
test['urdu_text'].fillna('', inplace=True)
```

# Tokenization

```python
import re

# Custom function to tokenize Urdu text using regex
def custom_word_tokenizer(text):
    # Regex pattern to split words and punctuation marks
    tokens = re.findall(r'\w+|[^\w\s]', text, re.UNICODE)
    return tokens

# Applying custom tokenizer to the train, test, and valid datasets
train['tokens'] = train['urdu_text'].apply(custom_word_tokenizer)
test['tokens'] = test['urdu_text'].apply(custom_word_tokenizer)
valid['tokens'] = valid['urdu_text'].apply(custom_word_tokenizer)

# Display tokenized text for the train dataframe as an example
print(train[['urdu_text', 'tokens']].head())
```

# TF-IDF 

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Re-join tokenized words back into sentences to use with TfidfVectorizer
train['joined_tokens'] = train['tokens'].apply(lambda x: ' '.join(x))
test['joined_tokens'] = test['tokens'].apply(lambda x: ' '.join(x))
valid['joined_tokens'] = valid['tokens'].apply(lambda x: ' '.join(x))

# Apply TF-IDF using the custom tokenizer
vectorizer = TfidfVectorizer(tokenizer=custom_word_tokenizer)

# Fit on train data and transform on all sets (train, test, valid)
tfidf_train = vectorizer.fit_transform(train['joined_tokens'])
tfidf_test = vectorizer.transform(test['joined_tokens'])
tfidf_valid = vectorizer.transform(valid['joined_tokens'])

# To see the TF-IDF scores for the train dataset
print("Top TF-IDF Scores for Train Dataset:")
feature_names = vectorizer.get_feature_names_out()
for doc_idx, doc in enumerate(tfidf_train):
    print(f"Document {doc_idx+1}:")
    for word_idx, score in zip(doc.indices, doc.data):
        print(f"Word: {feature_names[word_idx]}, Score: {score}")
```

# Phase 4 N gram Analysis

```python
import re
from collections import Counter
from nltk.util import ngrams



# Function to get n-grams
def get_ngrams(tokens_list, n):
    all_ngrams = []
    for tokens in tokens_list:
        all_ngrams.extend(list(ngrams(tokens, n)))
    return Counter(all_ngrams)

# Unigram, Bigram, and Trigram Analysis
def ngram_analysis(data, n):
    tokenized_data = data['tokens']
    ngram_freq = get_ngrams(tokenized_data, n)
    return ngram_freq.most_common(10)

# Get top 10 unigrams, bigrams, and trigrams for training dataset
top_10_unigrams = ngram_analysis(train, 1)
top_10_bigrams = ngram_analysis(train, 2)
top_10_trigrams = ngram_analysis(train, 3)

print("Top 10 Unigrams in Train Data:")
for unigram, count in top_10_unigrams:
    print(f"{unigram}: {count}")

print("\nTop 10 Bigrams in Train Data:")
for bigram, count in top_10_bigrams:
    print(f"{bigram}: {count}")

print("\nTop 10 Trigrams in Train Data:")
for trigram, count in top_10_trigrams:
    print(f"{trigram}: {count}")

# Perform analysis for the test and validation datasets similarly:
print("\nTop 10 Unigrams in Test Data:")
print(ngram_analysis(test, 1))

print("\nTop 10 Bigrams in Valid Data:")
print(ngram_analysis(valid, 2))
```

# Logistic regression

```python
import numpy as np  # Ensure numpy is imported
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
import gensim

# Assuming that the 'train', 'test', 'valid' datasets are already available and tokenized
# Let's use TF-IDF or Word2Vec for feature extraction

# Option 1: Using TF-IDF
vectorizer = TfidfVectorizer(tokenizer=custom_word_tokenizer)
X_train_tfidf = vectorizer.fit_transform(train['joined_tokens'])
X_test_tfidf = vectorizer.transform(test['joined_tokens'])
X_valid_tfidf = vectorizer.transform(valid['joined_tokens'])

# Option 2: Using Word2Vec
def get_word2vec_features(tokens, model, vector_size=100):
    # Averages the word vectors for each tokenized sentence
    vectors = []
    for token_list in tokens:
        vec = np.mean([model.wv[word] for word in token_list if word in model.wv] or [np.zeros(vector_size)], axis=0)
        vectors.append(vec)
    return np.array(vectors)

# Combine all tokenized data to train Word2Vec model
all_tokenized_data = train['tokens'].tolist() + test['tokens'].tolist() + valid['tokens'].tolist()
word2vec_model = gensim.models.Word2Vec(sentences=all_tokenized_data, vector_size=100, window=5, min_count=1, workers=4)

# Get Word2Vec features for train, test, and validation sets
X_train_w2v = get_word2vec_features(train['tokens'], word2vec_model)
X_test_w2v = get_word2vec_features(test['tokens'], word2vec_model)
X_valid_w2v = get_word2vec_features(valid['tokens'], word2vec_model)

# Using y_train, y_test, y_valid labels for the sentiment classification
y_train = train['is_sarcastic_encoded']
y_test = test['is_sarcastic_encoded']
y_valid = valid['is_sarcastic_encoded']

# Step 1: Building the model (you can choose one of these)
# Logistic Regression
log_reg = LogisticRegression(max_iter=200)
log_reg.fit(X_train_tfidf, y_train)  # Using TF-IDF features
# Or for Word2Vec: log_reg.fit(X_train_w2v, y_train)

# Step 2: Model Evaluation
# Predict on test data
y_pred = log_reg.predict(X_test_tfidf)  # Using TF-IDF features
# Or for Word2Vec: y_pred = log_reg.predict(X_test_w2v)

# Step 3: Calculating Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Displaying metrics
print("Model Performance:")
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

```

```python
import re

# Custom tokenizer function (the same one used for model training)
def custom_word_tokenizer(text):
    tokens = re.findall(r'\w+|[^\w\s]', text, re.UNICODE)
    return tokens

# Function to predict sarcasm
def predict_sarcasm(input_text, model, vectorizer):
    """
    Predict whether the given Urdu post is sarcastic or not.
    
    :param input_text: The input Urdu text to classify.
    :param model: The trained machine learning model (Logistic Regression, SVM, etc.).
    :param vectorizer: The fitted vectorizer (TF-IDF or Word2Vec) used for feature extraction.
    
    :return: The classification result: 'Sarcastic' or 'Non-Sarcastic'.
    """
    # Step 1: Preprocess the input (tokenize)
    tokens = custom_word_tokenizer(input_text)
    
    # Step 2: Re-join tokens to form a single string (as expected by the vectorizer)
    joined_tokens = ' '.join(tokens)
    
    # Step 3: Transform the input using the TF-IDF vectorizer (same as used in training)
    input_features = vectorizer.transform([joined_tokens])
    
    # Step 4: Predict using the trained model
    prediction = model.predict(input_features)
    
    # Step 5: Return the result (0 = Non-Sarcastic, 1 = Sarcastic)
    if prediction[0] == 1:
        return "Sarcastic"
    else:
        return "Non-Sarcastic"

# Example usage of the function

# Assuming 'log_reg' is the trained Logistic Regression model and 'vectorizer' is the fitted TF-IDF vectorizer

input_text = "دل بڑا کرو صاحب باتیں تو سب بڑی کرتے ہیں..🍂"  # Example Urdu input (likely sarcastic)

# Call the prediction function
result = predict_sarcasm(input_text, log_reg, vectorizer)

# Output the result
print(f"Prediction: {result}")
```

# 2. Challenges in Urdu Sentiment Analysis:
Performing sentiment analysis in Urdu presents a unique set of challenges:

### a. Complex Morphology:
Word Forms: Urdu is a morphologically rich language, where words change forms based on gender, number, tense, and case. This makes tokenization and handling word inflections difficult.
Compound Words: Some words in Urdu are composed of multiple components, which, if tokenized incorrectly, may result in loss of meaning.
 ### b. Colloquial and Informal Language:
Slang and Dialects: Urdu, especially on social media, is filled with regional variations, slang, and informal language that may not be present in formal language corpora.
Code-Switching: Urdu speakers frequently mix English and Urdu in informal communication. Detecting sentiment in such mixed-language texts can be challenging for traditional models that aren't trained on code-switched data.
 ### d. Contextual Sarcasm and Irony:
Implicit Meaning: Sarcasm and irony in Urdu often require an understanding of cultural and social context. Traditional models that rely on word frequency or simple patterns struggle with this implicit meaning.
